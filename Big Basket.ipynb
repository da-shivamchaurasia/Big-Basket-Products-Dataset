```python
# Find out Top &amp; least sold products
# Top sold product
top_market_price = df[df['market_price'] == df['market_price'].max()]
print("\nTop Market Price Product(s):")
print(top_market_price[['product','market_price']])

# Least sold product
least_market_price = df[df['market_price'] == df['market_price'].min()]
print("\nLeast Market Price Product(s):")
print(least_market_price[['product','market_price']])
```

```python
# Measuring discount on a certain item
df['discount_percentage'] = ((df['market_price'] - df['sale_price']) / df['market_price']) * 100
print("\nSample Discount Calculations:")
print(df[['product','market_price','sale_price','discount_percentage']].head(10))
```

```python
# Find out the Missing Values from the Dataset
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Find out the outliers and fill them with mean
numeric_cols = ['sale_price', 'market_price', 'rating', 'discount_percentage']

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    mean_value = df[col].mean()
    df[col] = np.where((df[col] < lower) | (df[col] > upper), mean_value, df[col])

print("\nOutliers handled successfully.")
```

```python
# Create Plots or visualizations

# Distribution of Sale Price
plt.figure(figsize=(8, 5))
sns.histplot(df['sale_price'], bins=30, kde=True)
plt.title('Distribution of Sale Price')
plt.show()
```

```python
# Discount vs Sale Price
plt.figure(figsize=(8, 5))
sns.scatterplot(x='discount_percentage', y='sale_price', data=df)
plt.title('Discount Percentage vs Sale Price')
plt.show()
```

```python
# Ratings Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['rating'], bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.show()
```
